## Data Manipulations


```python
import pandas as pd
```

```python
df = pd.read_csv('FSL.csv')

# df['Total'].head(5).mean()
```

```python
## Handling missing values 

df.isnull().any(axis=0)

print(df.isnull().sum())
```

```python
## fliing missing values

df.fillna(0)
```

```python
# filling missing values with the mean of the column

df['isNan'] = df['S2: Refugees and IDPs'].fillna(df['S2: Refugees and IDPs'].mean())
```

```python
df['S2: Refugees and IDPs'].head(40).isnull().sum()
```

```python
# renaming columns

df = df.rename(columns={'Country':'All Country'})
df.head(5)
```

```python
# changing data type
df['New Total'] = df['Total'].astype(int)
df.head(5)
```

```python
df.loc[0]
```

```python
grouped_sum = df.groupby(['Rank','Year'])['Total'].sum()
print(grouped_sum)
```

```python
# aggregate multiple functions
grouped_by_agg = df.groupby('All Country')['Total'].agg(['sum','mean','count'])
print(grouped_by_agg)
```

```python
## Merginh and Joining DataFrames

# Creating Sample DataFrame
df1 = pd.DataFrame({'Key':['A','B','C'],'Value1':[1,2,3]})
df2 = pd.DataFrame({'Key':['A','B','D'],'Value1':[4,5,6]})
```

```python
print(df1)
print(df2)
```

```python
# Merging DataFrame on the key column

df3 = pd.merge(df1,df2,on='Key',how='inner')
df3
```

```python
df3 = pd.merge(df1,df2,on='Key',how='outer')
df3
```
