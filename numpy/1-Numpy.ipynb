```python
## identity matrix
np.eye(3)
```

```python
### numpy vectorize 

ar1 = np.array([4,6,4,6,7])
ar2 = np.array([5,6,4,7,3])

## element wise addition 
sum = ar1+ar2
print(sum)
## element wise subsraction
sub = ar1-ar2
print(sub)
```

```python
## universal function
arr = np.array([4,6,3,16,9,81])

# square root 
print(np.sqrt(arr))

# exponential 
print(np.exp(arr))

# sine 
print(np.sin(arr))

# natural log
print(np.log(arr))
```

```python
# array slicing and indexing 
arr3 = np.array([[4,6,8],[4,5,3],[5,2,6]])

print(arr3[1:,1:2])
print(arr3[0][2])
```

```python
# statistical concept --normalization

arr4 = np.array([56,3,55,33,11,23,7,33])

## avarege 
print("mean :",np.mean(arr4))
#  median
print("median :",np.median(arr4))
## standard deviation
print("standard deivations :",np.std(arr4))
# variance
print("variance :",np.var(arr4))
```

```python
# logical operations

data = np.array([5,6,8,4,4,8,3])

result = data >=5
print(result)

get_data = data[data>=5]
print(get_data)
```
